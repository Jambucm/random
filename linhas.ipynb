```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def afim(x:list, a:float, b:float):
    y = []
    for i in x:
        y.append(a * i + b)
    return y
```

```python
a = [i for i in range(11)]
a = np.linspace(-1, 1, 11)
b = [j for j in range(11)]
x = [k for k in range(11)]
lista_y = []
print(a)
print(b)
for l in range(10):
    y = afim(x, a[l], b[l])
    lista_y.append(y)
```

```python

fig, ax = plt.subplots()
ax.axis(xmin=0, xmax=10, ymin=0, ymax=10)

for y in lista_y:
    plt.plot(x, y, color='black')

for y in lista_y:
    plt.plot(x[::-1], y, color='black')

plt.show()
```
